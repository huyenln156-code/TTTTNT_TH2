```python
##bài1

from google.colab import files

def upload_file():
    print("Hãy chọn file trên máy của bạn")
    uploaded = files.upload()
    ten_file = list(uploaded.keys())[0]
    return ten_file


def doc_ma_tran_ke_tu_file(ten_file):
    ma_tran = []
    with open(ten_file, "r") as f:
        for dong in f:
            if dong.strip():
                hang = list(map(int, dong.strip().split()))
                ma_tran.append(hang)
    return ma_tran


def tinh_bac_dinh(ma_tran_ke):
    return [sum(hang) for hang in ma_tran_ke]


def sap_xep_dinh_theo_bac(ds_dinh, bac):
    return [dinh for _, dinh in sorted(zip(bac, ds_dinh), reverse=True)]


def tao_bang_chi_muc(ds_dinh):
    return {ds_dinh[i]: i for i in range(len(ds_dinh))}


def to_mau_do_thi(ma_tran_ke, ds_dinh):
    chi_muc = tao_bang_chi_muc(ds_dinh)
    bac = tinh_bac_dinh(ma_tran_ke)
    ds_dinh_sap_xep = sap_xep_dinh_theo_bac(ds_dinh, bac)

    danh_sach_mau = ["Blue", "Red", "Yellow", "Green", "Orange", "Purple"]
    tap_mau = {dinh: danh_sach_mau.copy() for dinh in ds_dinh}
    ket_qua = {}

    for dinh in ds_dinh_sap_xep:
        mau_chon = tap_mau[dinh][0]
        ket_qua[dinh] = mau_chon

        hang_ke = ma_tran_ke[chi_muc[dinh]]
        for j in range(len(hang_ke)):
            if hang_ke[j] == 1:
                dinh_ke = ds_dinh[j]
                if mau_chon in tap_mau[dinh_ke]:
                    tap_mau[dinh_ke].remove(mau_chon)
    return ket_qua

ten_file = upload_file()

G = doc_ma_tran_ke_tu_file(ten_file)
so_dinh = len(G)
ds_dinh = [chr(ord('A') + i) for i in range(so_dinh)]

ket_qua = to_mau_do_thi(G, ds_dinh)

print("\nKẾT QUẢ TÔ MÀU ĐỒ THỊ:")
for dinh in sorted(ket_qua.keys()):
    print(f"Đỉnh {dinh}: {ket_qua[dinh]}")
```

```python
##Bài 3
from itertools import permutations
from google.colab import files

def upload_file():
    print("Hãy chọn file chứa ma trận:")
    uploaded = files.upload()
    return list(uploaded.keys())[0]

def doc_ma_tran_tu_file(ten_file):
    mat = []
    with open(ten_file, "r") as f:
        for dong in f:
            if dong.strip():
                hang = list(map(int, dong.split()))
                mat.append(hang)
    return mat

def tim_duong_di_min(cost):
    n = len(cost)
    duong_di_tot = None
    chi_phi_tot = float('inf')
    cac_lo_trinh = permutations(range(1, n))

    for lo_trinh in cac_lo_trinh:
        lo_trinh = (0,) + lo_trinh
        chi_phi = 0
        hop_le = True

        for i in range(len(lo_trinh) - 1):
            u = lo_trinh[i]
            v = lo_trinh[i+1]

            if cost[u][v] == 0:
                hop_le = False
                break

            chi_phi += cost[u][v]

        if hop_le and chi_phi < chi_phi_tot:
            chi_phi_tot = chi_phi
            duong_di_tot = lo_trinh

    return duong_di_tot, chi_phi_tot

ten_file = upload_file()
cost = doc_ma_tran_tu_file(ten_file)

duong_di, chi_phi = tim_duong_di_min(cost)

print("\nKẾT QUẢ ĐƯỜNG ĐI")
print("Đường đi tối ưu:", " → ".join([chr(ord('A') + i) for i in duong_di]))
print("Chi phí:", chi_phi)
```
